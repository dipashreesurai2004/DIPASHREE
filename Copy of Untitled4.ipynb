# Task
Create a heatmap of seasonal temperature variations using the data in "GlobalTemperatures.csv".

## Prepare data for heatmap

### Subtask:
Extract the month and year from the 'dt' column and calculate the average temperature for each month and year.


**Reasoning**:
Extract the month and year from the 'dt' column, group by year and month, calculate the mean of 'AverageTemperature', and then unstack to prepare for the heatmap.



```python
df_global_temp['month'] = df_global_temp['dt'].dt.month
df_global_temp['year'] = df_global_temp['dt'].dt.year
df_monthly_avg = df_global_temp.groupby(['year', 'month'])['AverageTemperature'].mean().unstack()
display(df_monthly_avg.head())
```

## Create the heatmap

### Subtask:
Generate a heatmap to visualize the seasonal temperature variations.


**Reasoning**:
Generate a heatmap to visualize the seasonal temperature variations as requested in the subtask instructions.



```python
plt.figure(figsize=(12, 8))
sns.heatmap(df_monthly_avg, cmap='viridis')
plt.title('Seasonal Temperature Variations Over Time')
plt.xlabel('Month')
plt.ylabel('Year')
plt.show()
```

## Summary:

### Data Analysis Key Findings

*   The average temperature for each month and year was calculated and structured into a DataFrame suitable for heatmap visualization.
*   A heatmap visualizing seasonal temperature variations over time was successfully generated, showing months on the x-axis and years on the y-axis, with colors representing the average temperature.

### Insights or Next Steps

*   The heatmap provides a visual representation of how average temperatures have varied across different months and years. Further analysis could involve examining specific patterns or trends visible in the heatmap, such as periods of unusually high or low temperatures.
*   Consider exploring the underlying data to identify potential drivers behind the observed temperature variations, such as climate change or specific weather events.
